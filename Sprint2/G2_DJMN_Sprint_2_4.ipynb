```python
# 1-Carregar o seu arquivo df_OVNI_limpo.csv (arquivo gerado após a limpeza de 
# dados efetuada na atividade 2.3) em um dataframe;

dataOVNIS = pd.read_csv('df_OVNI_limpo.csv')
dataOVNIS
```

```python
# 2-Dividir o conteúdo da coluna Date / Time em duas novas colunas no mesmo dataframe e deletar a coluna Date / Time .

# Dividir Data e Hora
dataOVNIS['Data'], dataOVNIS['Hora'] = dataOVNIS['Data/Hora'].str.split(' ',1).str
 
# Deletar coluna 'DateTime' e 'Unnamed: 0'
dataOVNIS.drop(columns=['Data/Hora'], inplace=True)
dataOVNIS.drop(columns=['Unnamed: 0'], inplace=True)

dataOVNIS
```

```python
# 3-Fazer o mesmo procedimento para dias da semana. Será que existe um dia da semana com mais 
# ocorrências de relatórios para OVNIs? Para descobrir isso, você deve criar uma nova coluna chamada weekdays.

# Extrair os dias da semana dos registro da coluna 'Data'
datas = pd.to_datetime(dataOVNIS['Data'])
datas = datas.dt.dayofweek

# Inserção dos dias da semana
dias_semana = {0: 'Segunda-feira',
               1:'Terça-feira',
               2: 'Quarta-feira',
               3: 'Quinta-feira',
               4: 'Sexta-feira',
               5: 'Sábado',
               6: 'Domingo'}

dataOVNIS['Dias_da_Semana'] = datas.map(dias_semana)

dataOVNIS
```

```python
# 4-Separar as variáveis mês (Month) e dia (Day). Desse modo, será possível refinar as pesquisas.

dataOVNIS['Dia'] = dataOVNIS['Data'].str.split('/', expand = True)[1]
dataOVNIS['Mês'] = dataOVNIS['Data'].str.split('/', expand = True)[0]

dataOVNIS
```

```python
# 5-Por fim, salvar o dataframe resultante em um arquivo .csv com o nome: 'df_OVNI_preparado'.

dataOVNIS.to_csv('df_OVNI_preparado.csv')
```
